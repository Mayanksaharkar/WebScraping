```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
response = requests.get('https://www.snapdeal.com/products/mens-footwear-sports-shoes?sort=plth')
```

```python
soap = BeautifulSoup(response.text, 'html.parser')

products = soap.find_all('div', attrs={'class': 'product-tuple-listing'})

data = []
print(type(data))

for product in products:
    product_title = product.find('p', attrs={'class': 'product-title'}).text
    # print(product_title)
    
    product_price = product.find('span', attrs={'class': 'product-price'}).text
    # print(product_price)
    
    product_image = product.find('img', attrs={'title': product_title}).attrs.get('data-src')
    # print(product_image)
    
    
    dict1 ={
        'product_title': product_title,
        'product_price': product_price,
        'product_image': product_image,
    }

    data.append(dict1)
    
# print(data)    
df = pd.DataFrame(data)   
df
```
